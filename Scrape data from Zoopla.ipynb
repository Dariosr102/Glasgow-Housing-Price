```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
lst1 = []
for i in range (0,2500,1):
    print(i)
    url = f"https://www.zoopla.co.uk/house-prices/glasgow/?q=glasgow&search_source=house-prices&numMonths=60&sortOrder=last_sale_date&sortDirection=desc&pn={i}"
    header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36"}
    page = requests.get(url, headers = header)
    soup = BeautifulSoup(page.content, 'lxml')
    main_data = soup.find_all("div",attrs={"class":"c-imWYEM c-imWYEM-pnwXf-hasUprn-true"})
    for data in main_data:
        try:
            title=data.find("h2", attrs={"class":"c-hWRJdU"}).get_text(strip=True)
        except AttributeError:
            title=np.nan
        try:
            date=data.find("label", attrs={"class":"c-eqprQi"}).get_text(strip=True)
        except AttributeError:
            date=np.nan
        try:
            price=data.find("label", attrs={"class":"c-cLIwII"}).get_text(strip=True)
        except AttributeError:
            price=np.nan
        try:
            bath=data.find("div", attrs={"class":"c-bkJvSo c-bkJvSo-egQFzo-isAnAttribute-true c-bkJvSo-iPJLV-css"}).get_text(strip=True).split("-")[0]
        except AttributeError:
            bath=np.nan
        try:
            bath=data.find("div", attrs={"class":"c-bkJvSo c-bkJvSo-egQFzo-isAnAttribute-true c-bkJvSo-iPJLV-css"}).get_text(strip=True).split("-")[0]
        except AttributeError:
            bath=np.nan
        try:
            todo=data.find("div", attrs={"class":"c-PJLV c-PJLV-iPJLV-css c-grgXa"}).get_text(", ", strip=True)
        except AttributeError:
            todo=np.nan
            
            
        lst1.append([title,date,price,todo])
```

```python
df=pd.DataFrame(data=lst1)
```
